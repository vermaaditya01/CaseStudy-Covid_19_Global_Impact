### 3) Unpivoting columns to turn tables from from wide to long format

##### a) confirmed Table

```python
conf_all_headers = confirmed.columns.tolist()
conf_date_headers = [h for h in conf_all_headers if h not in fixed_headers]
conf_date_headers
```

```python
## melting the table
confirmed_long = confirmed.melt(
    id_vars = fixed_headers, value_vars = conf_date_headers,
    var_name = "Date", value_name = "Confirmed"
)
confirmed_long
```

##### b) deaths Table

```python
death_all_headers = deaths.columns.tolist()
death_date_headers = [h for h in death_all_headers if h not in fixed_headers]
death_date_headers
```

##### c) recovered columns

```python
recover_all_headers = recovered.columns.tolist()
recover_date_headers = [h for h in recover_all_headers if h not in fixed_headers]
recover_date_headers
```

```python
## melting the table
recovered_long = recovered.melt(id_vars = fixed_headers, value_vars = recover_date_headers, var_name = "Date", value_name = "Recovered")
recovered_long
```

### 4) Handling missing values in each data table

##### a) confirmed_long:-

```python
confirmed_long.info()
```

```python
confirmed_long["Province/State"] = confirmed_long["Province/State"].fillna("All Provinces")

confirmed_long['Lat'] = pd.to_numeric(confirmed_long['Lat']).fillna(0)
confirmed_long['Long'] = pd.to_numeric(confirmed_long['Long']).fillna(0)

confirmed_long['Date'] = pd.to_datetime(confirmed_long['Date'], format = "%m/%d/%y", errors = "coerce")

confirmed_long.info()
```

```python
confirmed_long
```

##### b) deaths_long:-

```python
deaths_long.info()
```

```python
deaths_long['Province/State'] = deaths_long['Province/State'].fillna('All Provinces')

deaths_long['Lat'] = pd.to_numeric(deaths_long['Lat']).fillna(0)
deaths_long['Long'] = pd.to_numeric(deaths_long['Long']).fillna(0)

deaths_long['Date'] = pd.to_datetime(deaths_long['Date'], format = "%m/%d/%y", errors = 'coerce')

deaths_long['Deaths'] = pd.to_numeric(deaths_long['Deaths']).fillna(0).astype('int64')

deaths_long.info()
```

```python
deaths_long
```

##### c) recovered_long:-

```python
recovered_long.info()
```

```python
recovered_long['Province/State'] = recovered_long['Province/State'].fillna('All Provinces')

recovered_long['Lat'] = recovered_long['Lat'].fillna(0)
recovered_long['Long'] = recovered_long['Long'].fillna(0)

recovered_long['Date'] = pd.to_datetime(recovered_long['Date'], format = "%m/%d/%y", errors = 'coerce')

recovered_long['Recovered'] = recovered_long['Recovered'].fillna(0).astype('int64')

recovered_long.info()
```

### 5) Applying aggregations to each Table in order to merge them

##### a) confirmed_long -> confirmed_cases

```python
confirmed_cases = confirmed_long.groupby(['Country/Region', 'Date'])['Confirmed'].sum().reset_index()
confirmed_cases
```

##### b) deaths_long -> death_cases

```python
death_cases = deaths_long.groupby(['Country/Region', 'Date'])['Deaths'].sum().reset_index()
death_cases
```

##### c) recovered_long -> recovered_cases

```python
recovered_cases = recovered_long.groupby(['Country/Region', 'Date'])['Recovered'].sum().reset_index()
recovered_cases
```

### 6) Merging tables

```python
merged_data = pd.merge(confirmed_cases, death_cases, on = ['Country/Region', 'Date'], how = 'inner')
merged_data = pd.merge(merged_data, recovered_cases, on = ['Country/Region', 'Date'], how = 'inner').sort_values(['Country/Region', 'Date'])
merged_data
```

#### 3) Analyze the peak number of daily new cases in Germany, France and Italy. Which country experienced the highest single-day surge, and when did it occur?

```python
target_countries = ['Germany', 'France', 'Italy']
target_data = (
    merged_data[merged_data['Country/Region'].isin(target_countries)]
    .groupby(['Country/Region', 'Date'])['Confirmed'].sum().reset_index()
    .sort_values(['Country/Region', 'Date'])
)
target_data
```

```python
## adding a new column that contains daily new confirmed cases (Cumulative -> Day-Wise records)

target_data['DailyNewCases'] = target_data.groupby('Country/Region')['Confirmed'].diff()
target_data
```

```python
target_data[target_data['DailyNewCases'].isna()]
```

```python
## Handling NaN values of the DailyNewCases column by filling them with current-date confirmed case count.

target_data['DailyNewCases'].fillna(target_data['Confirmed'], inplace = True)
target_data
```

```python
# Creating a pivot table for a time-series matrix

daily_new_cases = target_data.pivot(index = 'Date', columns = 'Country/Region', values = 'DailyNewCases').sort_index()
daily_new_cases
```

```python
countrywise_cases = {}

# target_countries = ['Germany', 'France', 'Italy']

for country in target_countries:
    if country in daily_new_cases.columns:
        s = daily_new_cases[country]
        countrywise_cases[country] = s

countrywise_cases
```

```python
## Finding out the maximum single day cases for each target country
# peaks {Country}: (max_value, date)

peaks = {}
for country, series in countrywise_cases.items():
    max_val = int(series.max())
    max_index = series.idxmax()
    peaks[country] = (max_val, max_index.date())

peaks
```

```python
## Visualization (Line Chart)

plt.figure(figsize = (12, 6))
for country, series in countrywise_cases.items():
    plt.plot(series.index, series.values, label = country)

plt.title('Daily New Confirmed Cases for Germany, Italy, France', fontweight = 'bold')
plt.xlabel('Date', fontweight = 'bold')
plt.ylabel('Confirmed Cases', fontweight = 'bold')
plt.ylim(-200000, 200000)
plt.legend()
plt.show()
```

##### Final Insights:-

##### Among the analyzed countries France recorded the highest single-day surge in Confirmed cases with 117,900 cases recorded on 11 April 2021, significantly exceeding the peak daily cases counts of Germany(49,044 cases) & Italy(40,902 cases).

#### 4) Compare the recovery rates between Germany and Mexico as of January 31, 2021. Which country showed better management of the pandemic according to this metric?

#### 9) Analyze monthly sum of confirmed cases, deaths and recoveries for countries to understand the progression of the pandemic.

```python
m = merged_data.copy()
m = m.sort_values(['Country/Region', 'Date'])
m['YearMonth'] = m['Date'].dt.to_period('M')
m
```

```python
# Daily Confirmed
m['DailyConfirmed'] = m.groupby('Country/Region')['Confirmed'].diff().clip(lower = 0).fillna(0)

# Daily Deaths
m['DailyDeaths'] = m.groupby('Country/Region')['Deaths'].diff().clip(lower = 0).fillna(0)

# Daily Recovered
m['DailyRecovered'] = m.groupby('Country/Region')['Recovered'].diff().clip(lower = 0).fillna(0)

m
```

```python
## Finding the monthly trends

monthly_trends = (
    m.groupby(['Country/Region', 'YearMonth'])[['DailyConfirmed', 'DailyDeaths', 'DailyRecovered']]
    .sum().rename(
        columns = {
            'DailyConfirmed' : 'Confirmed',
            'DailyDeaths' : 'Deaths',
            'DailyRecovered' : 'Recovered',
        }
    ).astype(int).reset_index().sort_values(['Country/Region', 'YearMonth'])
)
monthly_trends
```
